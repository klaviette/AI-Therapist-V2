```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array
from PIL import Image
import os

datagen = ImageDataGenerator(
        rotation_range = 40, 
        width_shift_range = 0.1,
        height_shift_range = 0.1,
        shear_range = 0.2,
        zoom_range = 0.1, 
        horizontal_flip = True,
        fill_mode = 'nearest')

DATADIR = "/kaggle/input/trainingdata/TrainingData"
target_dir = "/kaggle/working/"
CATEGORIES = ['Happy', 'Sad']

for category in CATEGORIES:
    path = os.path.join(DATADIR, category)  # path to our files
    os.makedirs(os.path.join(target_dir, category), exist_ok=True)
    for img_name in os.listdir(path):
        img_path = os.path.join(path, img_name)
        
        # Load the image using PIL
        img = Image.open(img_path)
        
        # Convert the PIL image to a NumPy array
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)
        
        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=target_dir + category, save_prefix='aug', save_format='jpg'):
            i += 1
            if i > 10:
                break
```
